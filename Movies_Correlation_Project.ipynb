```python
# import libraries

import pandas as pd
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt

```

```python
# Read data
df = pd.read_csv('movies.csv')
```

```python
df = df.dropna()
```

```python
# Change data type of column

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
# Extracting year from the released date

df['released_year'] = df['released'].str.extract(r'(\d{4})')
```

```python
df.head()
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Dropping if there is duplicates
df.drop_duplicates()
```

```python
# Checking correlations

# Scatter Plot with budget vs gross
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget for Film')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
# Plot with regression line

sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color':'red'}, line_kws={'color':'blue'})
```

```python
# Correlation between numerical columns
correlation = df.corr(numeric_only=True)
sns.heatmap(correlation, annot=True)
plt.title('Correlation Matrix')
plt.show()




```

```python
df_numerized = df
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
df_numerized
```

```python
# Correlation between numerical columns
correlation = df_numerized.corr(numeric_only=True)
plt.figure(figsize=(12, 8))
sns.heatmap(correlation, annot=True,cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()




```

```python
# Filtering out the correlation more than 0.5
corr_mat = df_numerized.corr(numeric_only=True)
corr_pairs = corr_mat.unstack()
sorted_pairs = corr_pairs.sort_values()
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
